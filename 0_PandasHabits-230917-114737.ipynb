```python
def clean_df(data_frame):
    temp_data = data_frame.copy()
    temp_data['make'] = temp_data.make.str.upper()
    temp_data['age'] = pd.Timestamp('today').year - temp_data['year']
    temp_data['city08'] = temp_data['city08'] / temp_data['city08'].mean()
    return temp_data

new_df = clean_df(new_df)
old_df = clean_df(old_df)
```
