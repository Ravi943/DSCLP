### **6.2 Handling of Missing Data**

- In this section we will see if we can **fill** the missing data with some **logical imputation.**

```python
# Finding Missing Data in the Dataset

null_frame = pd.DataFrame(index = df.columns.values)
null_frame['Null Frequency'] = df.isnull().sum().values

percent = df.isnull().sum().values/df.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100

null_frame
```

**Observation:**

- **Feature:**
  - Problem &rarr; Action Required

- **Gender:**

  - **Missing Values (116)** &rarr; **Impute with Mode**

- **City_Category:**

  - **Missing Values (72)** &rarr; **Impute with Mode**

- **Product_Category:**

  - **Missing Values(160)** &rarr; **Either Fill using Product ID or Drop Rows**

```python
# Imputing all the Missing Values in Gender and City_Category

# Gender
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

# City_Category
df['City_Category'].fillna(df['City_Category'].mode()[0], inplace=True)
```

```python
# Cross-Checking the Operations
df.isnull().sum()
```

**Observation:**

- Missing Values in the **Gender and City_category** has been **Filled Successfully.**

- Now, to fill the missing values in **Product_Category**:

    - We have **Product ID** in the dataset, so we can use that information to fill the missing values in **Product_Category.**

    - First create a **Dictionary** that **maps** every **Product_ID** to it\'s **corresponding Product_Category.**

    - Use that **Dictionary to fill the missing Product_Category** if that product was ordered in any of the rest of the purchases.

    - Finally **drop any rows** in which **we can not fill Product_Category** using the above described method.

```python
# Imputing Missing Values in Product_Category

# Creating the Dictionary
mapper = {}

for row in range(df.shape[0]):
    ID = df.loc[row, 'ProductID']
    Category = df.loc[row, 'Product_Category']
    mapper[ID] = mapper.get(ID, Category)

# Printing the top 8 ProductID and Product_Category Pair
for ID, Category in list(mapper.items())[:8]:
    print('ProductID:', ID, ' Product_Category:', Category)
```

```python
# Trying to Imput the Values using the mapper Dictionary
indexes = df[df['Product_Category'].isnull()].index

for index in indexes:
  Category = mapper[df.loc[index, 'ProductID']]
  df.loc[index, 'Product_Category'] = Category
```

```python
# Checking the Missing Values
df.isnull().sum()
```

**Observation:**

- All the **Missing values in the Product Category has been filled** using the information of ProductID

**<h4>Question 12: Which Product Categories are sold more on average based on the gender of the customer? </h4>**
