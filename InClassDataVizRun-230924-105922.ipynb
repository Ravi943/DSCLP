```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
# data[]   # SQUARE BRACKETS TO EXTRACT A COLUMN FROM DATAFRAME
# function()   # normal brackets / paranthesis
# {}  # dictionaries / sets
```

## 1. SALES TREND

```python
daily_sales = data.groupby("Order Date").agg({"Sales": "sum"})  # DAILY SALES
monthly_sales = daily_sales.resample('M').sum()    # MONTHLY SALES
```

```python
fig, my_axis = plt.subplots(figsize=(14, 5))

monthly_sales.plot(kind='line', ax=my_axis, marker='o')

plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

plt.grid(linestyle=':')
plt.tight_layout()
```

## 2. PROFITS TREND

```python
daily_profits = data.groupby('Order Date').agg({'Profit': "sum"})
monthly_profits = daily_profits.resample('M').sum()
```

```python
fig, my_axis = plt.subplots(figsize=(14, 5))

monthly_profits.plot(kind='line', ax=my_axis, marker='o')

plt.ylabel("Profits")
plt.title("Monthly Profits Trend")

plt.grid(linestyle=':')
plt.tight_layout()
```

### 3. Quarterly SALES & PROFIT AS A BAR CHART

```python
daily_data = data.groupby('Order Date').agg({"Sales": "sum", "Profit": "sum"})
monthly_data = daily_data.resample('Q').sum()

monthly_data.index = monthly_data.index.strftime("%b-%Y")  # strftime.org
```

```python
fig, my_axis = plt.subplots(figsize=(14, 6))

monthly_data.plot(kind='bar', ax=my_axis)
plt.title("Quarterly Sales & Profit Trend")
plt.ylabel("Amount ($)")

plt.grid(linestyle=':')

plt.tight_layout()
```

### STACKED BAR CHART - REGIONAL SALES PER QUARTER

```python
daily_region_data = (data
                     .groupby(["Region", "Order Date"])
                     .agg({"Sales": "sum"})
                     .unstack(0))

q_region_data = (daily_region_data
                 .resample('Q')
                 .sum())

q_region_data.index = q_region_data.index.strftime('%b-%Y')
q_region_data = q_region_data.droplevel(axis=1, level=0)
```

```python
fig, my_axis = plt.subplots(figsize=(14, 5))

q_region_data.plot(kind='bar', stacked=True, ax=my_axis)

plt.ylabel("Amount ($)")
plt.xlabel(None)
plt.title("Region wise sales per Quarter")

plt.tight_layout()
```
